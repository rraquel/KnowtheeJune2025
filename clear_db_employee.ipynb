```python
# clear_employee_table.py

from sqlalchemy import create_engine, text, inspect

DATABASE_URL = "postgresql://postgres:postgres@127.0.0.1:5433/knowthee"
engine = create_engine(DATABASE_URL)

with engine.connect() as connection:
    inspector = inspect(engine)
    tables = inspector.get_table_names()
    if "employees" not in tables:
        print("❌ Table 'employee' does not exist. Available tables:", tables)
    else:
        connection.execute(text("DELETE FROM employees;"))
        connection.commit()
        print("✅ Cleared all rows from 'employee' table.")
```
